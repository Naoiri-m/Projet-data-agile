## Préparation des données sources

### Importation

```python
df_infopro = pd.read_excel('./datasets/table-info-pro.xlsx')
df_remuneration = pd.read_excel('./datasets/table-remuneration.xlsx')
df_salaries = pd.read_excel('./datasets/table-salaries.xlsx')

dfs = ['df_infopro', 'df_remuneration', 'df_salaries']

for df in dfs:
    display(eval(df).head())
del df
```
